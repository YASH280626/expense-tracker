yearly_summary is now a dataframe

```python
yearly_summary = df.groupby([df['Date'].dt.year, 'Type'])['Amount'].sum().unstack()
print(yearly_summary)
```

```python
yearly_summary['Savings'] = yearly_summary['Income'] - yearly_summary['Expense']
print(yearly_summary)
```

creating a copy of df to change salary from expense to income 

```python
clean_df = df.copy()
clean_df.loc[clean_df['Category'] == 'Salary', 'Type'] = 'Income'
clean_df.head()
```

```python
clean_df.groupby('Category')['Type'].unique()
```

```python
clean_yearly_summary = clean_df.groupby([clean_df['Date'].dt.year, 'Type'])['Amount'].sum().unstack()
```

```python
clean_yearly_summary
```

```python
clean_yearly_summary['Savings'] = clean_yearly_summary['Income'] - clean_yearly_summary['Expense']
print(clean_yearly_summary)
```

```python
expensive_category = clean_df[clean_df['Type'] == 'Expense'].groupby('Category')['Amount'].sum().sort_values(ascending=False)
print(expensive_category)
```

```python
expensive_df = clean_df[clean_df['Type'] == 'Expense']
expensive_year = expensive_df[expensive_df['Type'] == 'Expense'].groupby([expensive_df['Date'].dt.year, 'Category'])['Amount'].sum()
print(expensive_year)
```

```python
expensive_by_year = expensive_year.unstack()
print(expensive_by_year)
```

```python
expensive_by_year.idxmax(axis = 1)
```

```python
expensive_by_year.max(axis=1)
```

```python
index_top_year = expensive_by_year.index
final_frame = pd.DataFrame(index=index_top_year)
final_frame['Top Categories'] = expensive_by_year.idxmax(axis=1)
final_frame['Amount'] = expensive_by_year.max(axis=1)
print(final_frame)
```
