yearly_summary is now a dataframe

```python
yearly_summary = df.groupby([df['Date'].dt.year, 'Type'])['Amount'].sum().unstack()
print(yearly_summary)
```

```python
yearly_summary['Savings'] = yearly_summary['Income'] - yearly_summary['Expense']
print(yearly_summary)
```

creating a copy of df to change salary from expense to income 

```python
clean_df = df.copy()
clean_df.loc[clean_df['Category'] == 'Salary', 'Type'] = 'Income'
clean_df.head()
```

```python
clean_df.groupby('Category')['Type'].unique()
```

```python
clean_yearly_summary = clean_df.groupby([clean_df['Date'].dt.year, 'Type'])['Amount'].sum().unstack()
```

```python
clean_yearly_summary
```

```python
clean_yearly_summary['Savings'] = clean_yearly_summary['Income'] - clean_yearly_summary['Expense']
print(clean_yearly_summary)
```
