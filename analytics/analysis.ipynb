creating a copy of df to change salary from expense to income 

```python
clean_df = df.copy()
clean_df.loc[clean_df['Category'] == 'Salary', 'Type'] = 'Income'
clean_df.head()
```

```python
clean_df.groupby('Category')['Type'].unique()
```

```python
clean_yearly_summary = clean_df.groupby([clean_df['Date'].dt.year, 'Type'])['Amount'].sum().unstack()
```

```python
clean_yearly_summary
```

```python
clean_yearly_summary['Savings'] = clean_yearly_summary['Income'] - clean_yearly_summary['Expense']
print(clean_yearly_summary)
```

```python
expensive_category = clean_df[clean_df['Type'] == 'Expense'].groupby('Category')['Amount'].sum().sort_values(ascending=False)
print(expensive_category)
```

```python
expensive_df = clean_df[clean_df['Type'] == 'Expense']
expensive_year = expensive_df[expensive_df['Type'] == 'Expense'].groupby([expensive_df['Date'].dt.year, 'Category'])['Amount'].sum()
print(expensive_year)
```

```python
expensive_by_year = expensive_year.unstack()
print(expensive_by_year)
```

```python
expensive_by_year.idxmax(axis = 1)
```

```python
expensive_by_year.max(axis=1)
```

```python
index_top_year = expensive_by_year.index
final_frame = pd.DataFrame(index=index_top_year)
final_frame['Top Categories'] = expensive_by_year.idxmax(axis=1)
final_frame['Amount'] = expensive_by_year.max(axis=1)
print(final_frame)
```

```python
expense_df = clean_df[clean_df['Type'] == 'Expense']
expense_df.head()
```

```python
expense_df['YearMonth'] = expense_df['Date'].dt.to_period('M')
```

```python
expense_df.head()
```

```python
monthly_expense = expense_df.groupby('YearMonth')['Amount'].sum()
```

```python
monthly_expense.plot(figsize=(12, 5))
plt.title('Monthly Expense Trend')
plt.xlabel('Month')
plt.ylabel('Expense')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
income_df = clean_df[clean_df['Type'] == 'Income']
income_df.head()
```

```python
income_df['YearMonth'] = income_df['Date'].dt.to_period('M')
income_df.head()
```

```python
monthly_income = income_df.groupby('YearMonth')['Amount'].sum()
monthly_income.head()
```

```python
monthly_income.plot(figsize=(12, 5))
plt.title('Monthly Income Trend')
plt.xlabel('Month')
plt.ylabel('Expense')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
monthly_expense.plot(figsize=(12,5), label='Expense')
monthly_income.plot(label='Income')
plt.title("Monthly Income vs Expense")
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.tight_layout()
plt.show()
```
